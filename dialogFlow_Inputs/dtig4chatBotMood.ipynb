#To work more on the Mood, i want to define a threshold for the mood to know if it is Happy (High or positve, good feelings songs), Lonely( middle, not sad, not lonely just there, Sad( Low, negative, Mourning, bitter)

```python
energy_threshold = 0.7
valence_threshold = 0.5
danceability_threshold = 0.6
```

```python
def assign_mood(row):
    if row['energy'] >= energy_threshold and row['valence'] >= valence_threshold and row['danceability'] >= danceability_threshold:
        return 'celebration'
    elif row['energy'] <= energy_threshold and row['valence'] <= valence_threshold and row['danceability'] <= danceability_threshold:
        return 'sad'
    elif row['energy'] >= energy_threshold and row['valence'] <= valence_threshold and row['danceability'] <= danceability_threshold:
        return 'lonely'
    # Add more conditions for other mood types
```

```python
df_mood = pd.read_csv('/content/drive/MyDrive/df_comb_v4.csv')
```

```python
df_mood['mood'] = df_mood.apply(assign_mood, axis=1)  # Apply the assign_mood function row-wise
```
